# Loading fbref player CSVs from a league

```python
import pandas as pd
```

```python
misc = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_miscelaneous_fbref.csv")
adgk = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_advancedgoalkeeping_fbref.csv")
defe = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_defensive_fbref.csv")
gk = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_goalkeeping_fbref.csv")
gsca = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_gsca_fbref.csv")
pas = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_passing_fbref.csv")
pastype = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_passtypes_fbref.csv")
pos = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_possesion_fbref.csv")
sho = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_shooting_fbref.csv")
stand = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_standard_fbref.csv")
time = pd.read_csv("/content/drive/MyDrive/AI_projects/LaLiga_players/laliga_2021_time_fbref.csv")

data = [misc, adgk, gk, defe, gsca, pas, pastype, pos, sho, stand, time];
```

```python
def name_editor(name):
  for i in range(len(name)):
    if name[i] == "\\":
      new_name = name[i+1:]
      break
  return new_name


for dataset in data:
  dataset.drop('Matches', axis='columns', inplace=True)
  for i in range(len(dataset)):
    name = name_editor(dataset.Player[i])
    dataset.at[i, "Player"] = name
```
